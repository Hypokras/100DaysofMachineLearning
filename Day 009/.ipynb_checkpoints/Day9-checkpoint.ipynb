#### Interecting with Database

```python
import sqlite3
```

```python
query = """
    CREATE TABLE example_db
    (name VARCHAR(20), age INTEGER);
"""
```

```python
# establish connection
con = sqlite3.connect('mydata.sqlite')
```

```python
#execute query
con.execute(query)
```

```python
con.commit()
```

```python
#insert rows
statment1 = "INSERT INTO example_db VALUES('Durgesh', 27)"
con.execute(statment1)
con.commit()
```

```python
# add many row at once

rows_value = [("ABC", 25), ("XYZ", 20), ("John", 30)]
statement2 = "INSERT INTO example_db VALUES(?,?)"
con.executemany(statement2, rows_value)
```

```python
con.commit()
```

```python
#get all values from database
statment3 = con.execute('select * from example_db')
rows = statment3.fetchall()
```

```python
rows
```

```python
statment3.description
```

```python
names = pd.DataFrame(rows, columns=[x[0] for x in statment3.description])
names
```
